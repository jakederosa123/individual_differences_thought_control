```python
import pandas as pd
from functools import reduce
    
def binnit(data, kind):
    
    def get_bins(data, col, kind):

        binned = (data[col].sort_values().reset_index()
                  .assign(binned = np.repeat(np.arange(1, 19), 20)))

        binned.columns = ['parcel', f'{kind}_{col}', f'{kind}_{col}_binned']
        #binned = binned.reset_index()
        return binned

    grad_bins = []
    for i in ['g1', 'g2', 'g3']:
        grad_bins.append(get_bins(data, i, kind))



    # Merge all DataFrames in the list on the 'parcel' column
    merged_df = reduce(lambda left, right: pd.merge(left, right, on='parcel'), grad_bins).sort_values('parcel')

    return merged_df

#RSA
rsa_task_bins = []
for i,j in zip([rsa_maintain_grads, rsa_replace_grads, rsa_suppress_grads, rsa_clear_grads],
             ['main', 'replace', 'suppress', 'clear']):
    rsa_task_bins.append(binnit(i, j))
rsa_bin_df = reduce(lambda left, right: pd.merge(left, right, on='parcel'), rsa_task_bins).sort_values('parcel')

#FC Task
fc_task_bins = []
for i,j in zip([fc_maintain_grads, fc_replace_grads, fc_suppress_grads, fc_clear_grads],
             ['main', 'replace', 'suppress', 'clear']):
    fc_task_bins.append(binnit(i, j))
    
fc_task_df = reduce(lambda left, right: pd.merge(left, right, on='parcel'), fc_task_bins).sort_values('parcel')

#FC Rest
fc_rest_bins = []
for i,j in zip([rest_grads], ['rest']):
    fc_rest_bins.append(binnit(i, j))
    
fc_rest_df = reduce(lambda left, right: pd.merge(left, right, on='parcel'), fc_rest_bins).sort_values('parcel')
```

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import mahalanobis

# Filter the DataFrame to include relevant columns
df = rsa_bin_df.filter(regex='parc|g(?!.*bin)')

# Function to calculate Mahalanobis distance
def calculate_mahalanobis(df, col1, col2):
    diff = df[col1].values - df[col2].values
    cov_matrix = np.cov(df[col1].values.T)
    inv_cov_matrix = np.linalg.inv(cov_matrix)
    m_distance = [mahalanobis(diff[i], np.zeros(diff.shape[1]), inv_cov_matrix) for i in range(diff.shape[0])]
    return m_distance

# List of operation pairs to calculate Mahalanobis distance
operation_pairs = [
    (['main_g1', 'main_g2', 'main_g3'], ['replace_g1', 'replace_g2', 'replace_g3']),
    (['main_g1', 'main_g2', 'main_g3'], ['suppress_g1', 'suppress_g2', 'suppress_g3']),
    (['main_g1', 'main_g2', 'main_g3'], ['clear_g1', 'clear_g2', 'clear_g3']),
    (['replace_g1', 'replace_g2', 'replace_g3'], ['suppress_g1', 'suppress_g2', 'suppress_g3']),
    (['replace_g1', 'replace_g2', 'replace_g3'], ['clear_g1', 'clear_g2', 'clear_g3']),
    (['suppress_g1', 'suppress_g2', 'suppress_g3'], ['clear_g1', 'clear_g2', 'clear_g3'])
]

# Calculate Mahalanobis distance for each pair of operations and each parcel
distances = {}
for col1, col2 in operation_pairs:
    pair_name = f"{col1[0].split('_')[0]}_vs_{col2[0].split('_')[0]}"
    distances[pair_name] = calculate_mahalanobis(df, col1, col2)

# Convert distances to a DataFrame
distances_df = pd.DataFrame(distances)

# Add parcel column to the distances DataFrame
distances_df['parcel'] = df['parcel']

# Summarize changes for each parcel
summary_df = distances_df.set_index('parcel')



# Display the summary DataFrame
summary_df = pd.merge(summary_df, glasser_regions, on='parcel')

# Save the summary DataFrame to CSV for further analysis
summary_df.to_csv('rsa_gradient_parcel_changes.csv')
```
